Question1: Backpropagation
In the following Neural Network, we have initialized the weights randomly. Use a training sample
(X,y) = ((1,1), 0) to update the weights (perform one round of backpropagation using one training
sample). Use learning rate parameter α = 0.1. Note that we have bias terms with value of 1 in
this network (no need for coding for this question).

```python
from numpy import array, exp 
x1 = 1
x2 = 1
y = 0
a1 = (1*.8) + (-0.5 * x1) + (.4 * x2)
a1 = 1 / (1 + exp(-(a1)))
a2 = (-0.1 * x2) + (0.9 * x2) + (1 * x2)
a2 = 1 / (1 + exp(-(a2)))
a_one = (1 * .3) + (a1 * -1.2) + (a2*1.1)
a_one =  1 / (1 + exp(-(a_one)))
output = a_one
delta = (y - output) * output * (1 - output)
print("a1: " + str(a1))
print("a2: " + str(a2))
print("output: " + str(output))
print("delta: " + str(delta))
```

```python
delta_one = (-1.2*(delta))*a1*(1.0-a1)
delta_two = (1.1*(delta))*a2*(1.0-a2)
print("delta_one: " + str(delta_one))
print("delta_two: " + str(delta_two))
```

```python
alpha = 0.1
weight_one_one = -1.2 + alpha * (delta)*  a1
weight_one_two = 1.1 + alpha * (delta)*  a2
weight_one_x1 = -0.5 + alpha * (delta_one) * 1
weight_two_x1 = 0.4 + alpha * (delta_one) * 1
weight_one_x2 = 0.9 + alpha * (delta_two) * 1
weight_two_x2 = 1.0 + alpha * (delta_two) * 1
print("weight_one_one: " + str(weight_one_one))
print("weight_one_two: " + str(weight_one_two))
print("weight_one_x1: " + str(weight_one_x1))
print("weight_two_x1: " + str(weight_two_x1))
print("weight_one_x2: " + str(weight_one_x2))
print("weight_two_x2: " + str(weight_two_x2))
```

b- Build the feature matrix and label vector: Each image is considered as a data sample with
pixels as features. Thus, to build the feature table you have to convert each 8x8 image
into a row of the feature matrix with 64 feature columns for 64 pixels

```python
# "read_csv" is a pandas function to read csv files from web or local device:
import os
label = pd.read_csv('label.csv')
y_label = label['digit']
        
total_df = pd.DataFrame()
directory = 'Digit/'
row_order_list = []
for file in os.listdir(directory):
    if file.endswith(".jpg"):
        image_num = file.split(".")[0]
        jpg = mpimg.imread(directory+file)
        jpg_df = pd.DataFrame(jpg).transpose()
#         print(file.split(".")[0])
        row_order_list.append(int(file.split(".")[0]))
        col_df = pd.DataFrame()
        for num in range(0, len(jpg_df)):
            temp_df = pd.DataFrame()
            array = pd.Series(jpg_df[num])
            temp_df[str(image_num)] = array
            col_df = col_df.append(temp_df)
            if num > 7:
                break
        total_df = total_df.append(col_df.transpose())

x_df = total_df
x_df['index'] = row_order_list
sorted_row_df =x_df.sort_values(['index'], ascending=[True])
sorted_row_df = sorted_row_df.drop('index', axis=1)
```

```python
sorted_row_df
```

c- Use sklearn functions to split the dataset into testing and training sets with the following
parameters: test_size=0.1, random_state=2

```python

X_train, X_test, y_train, y_test = train_test_split(sorted_row_df, y_label, test_size=0.1, random_state=2)
```

d- Design and Train an ANN with one hidden layer with 80 neurons to recognize the digits
based on the training dataset that you built in part (c). Use random_state=1,
learning_rate_init = 0.002. Then, Test your ANN on testing set (from part(c)), and
calculate and report the accuracy. Also, calculate and report the Confusion Matrix.

```python
my_ANN = MLPClassifier(hidden_layer_sizes=(80,), activation= 'logistic', 
                       solver='adam', alpha=1e-5, random_state=1, 
                       learning_rate_init = 0.002)

my_ANN.fit(X_train, y_train)
```

```python
y_predict_ann = my_ANN.predict(X_test)
score_ann = accuracy_score(y_test, y_predict_ann)
print('ANN Accuracy: ' + str(score_ann))
```

```python
cm_ANN = metrics.confusion_matrix(y_test, y_predict_ann)
print("Confusion matrix:")
print(cm_ANN)
```

```python
# Searching for the best number of neurons by using GridSearch

neuron_number = [(i,) for i in range(50,200)]

param_grid = dict(hidden_layer_sizes = neuron_number)

my_ANN = MLPClassifier(activation='logistic', solver='adam', alpha=1e-5, random_state=1, learning_rate_init = 0.002)

grid = GridSearchCV(my_ANN, param_grid, cv=10, scoring='accuracy')

grid.fit(sorted_row_df, y_label)

print('Grid Best Accuracy Score: ' + str(grid.best_score_))
print('Grid Best Parameters: ' + str(grid.best_params_))
```
